### Creating a quantum object
A quantum object can be quantum states or quantum operators

##### We can create a quantum object by using qx() as follows:


```python
a = qx([[1],[0]])
a
```


##### We can create some standard quantum states by calling their names.
<br>
Such as, below is a random state 

```python
x = random(2)
x
```

List of standard states in tqix can be seen here:

##### We can check the properties of a quatum object as follows

```python
# data type
print(typex(x))
```

```python
# the dimension, or the Hilbert state space
print(shapex(x))
```

```python
# check if x is quantum object or not
isqx(x)
```

```python
# check if x is normalized or not
isnormx(x)
```

```python
# convert to an operator
ox = operx(x)
ox
```

```python
# check if ox is Hermit or not
ishermx(ox)
```

### Common operated functions on quantum objects

```python
# convert a bra or ket vector into oper
x = ghz(2)
ox = operx(x)
```

```python
# diagonalize matrix ox
diagx(ox)
```

```python
# get conjugate transpose of x: x^†
daggx(x)
```

```python
# get conjugation of x: x^∗
conjx(x)
```

```python
# get transpose of x: x^T
transx(x)
```

```python
# get trace of ox: (only for oper)
tracex(ox)
```

```python
# eigenvalue and eigenstate
eigenx(ox)
```

```python
# get ground state for a given Hamiltonian
groundx(ox)
```

```python
# exponentiated ox
expx(ox)
```

```python
# square root of ox
sqrtx(ox)
```

```python
# getnorm2 of x
l2normx(x)
```

```python
# get normalize of x
normx(x)
```
